```python
%matplotlib notebook
g = sns.PairGrid(college_data, height = 2.5)
g  = g.map(plt.scatter)

# Setting the labels of the axis in each grid
x_labels, y_labels = [],[]
for ax in g.axes[-1,:]:
    xlabel = ax.xaxis.get_label_text()
    x_labels.append(xlabel)
for ax in g.axes[:,0]:
    ylabel = ax.yaxis.get_label_text()
    y_labels.append(ylabel)

for i in range(len(x_labels)):
    for j in range(len(y_labels)):
        g.axes[j,i].xaxis.set_label_text(x_labels[i])
        g.axes[j,i].yaxis.set_label_text(y_labels[j])
```

## K Means Cluster Creation

**Find the n_cluster using the Elbow Method**

```python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
```

```python
x = college_data[['F.Undergrad', 'Outstate']]
```

```python
zipped_x = np.array(list(zip(np.array(college_data['F.Undergrad']), np.array(college_data['Outstate']))))
```

```python
distortions = []
for i in range (1, 6):
    model = KMeans(n_clusters = i)
    model.fit(x)
    # Get the minimum eucldistance of each data point from the each center and store them into the distortions.
    distortion = sum(np.min(cdist(zipped_x, model.cluster_centers_, 'euclidean'), axis = 1))
    distortions.append(distortion)
```

**We will plot the graph to check the best n_clusters value**

```python
plt.figure(figsize = (10, 6))
plt.plot(range(1, 6), distortions, 'bo-')
plt.xlabel('Value of K')
plt.ylabel('Distortion')
```

** We can see the best value of the n_clusters is 3. Now we can use to fit the model and add one more column as 'cluster.result' **

```python
model = KMeans(n_clusters = 3)
model.fit(x)
college_data['cluster.result'] = model.labels_
```
